## Questions 
1. Who are the top 10 players of the match?
2. How many wins occurred when batting first?

```python
# load the Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# remove warnings 
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset 
df = pd.read_csv("matches.csv")
```

```python
# show the first 5 rows
df.head()
```

```python
# show the shape
df.shape
```

```python
# find player of the match
df['player_of_match'].value_counts()
```

```python
# top 10 player of the match
df['player_of_match'].value_counts()[:10]
```

```python
# top 5 player of the match
df['player_of_match'].value_counts()[:5]
```

```python
list(df['player_of_match'].value_counts()[:5].keys())
```

```python
# make a bar plot of top five players
plt.figure(figsize=(10,5))
colors=['green','red','blue','yellow','black']
plt.bar(list(df['player_of_match'].value_counts()[:5].keys()),list(df['player_of_match'].value_counts()[:5]),color=colors)
plt.show()
```

```python
# making a  bar plot of top 10 players
plt.figure(figsize=(12,5))
plt.bar(list(df['player_of_match'].value_counts()[:10].keys()),list(df['player_of_match'].value_counts()[:10]),color="g")
plt.show()
```

```python
# frequency of result column
df['result'].value_counts()
```

```python
# finding out the number of toss w.r.t win team
df['toss_winner'].value_counts()
```

```python
# find the record where team won by batting first
bf=df[df['win_by_runs']!=0]
```

```python
bf.head()
```

```python
# making a histogram
plt.figure(figsize=(5,5))
plt.hist(bf['win_by_runs'])
plt.title("Distribution Of runs")
plt.xlabel("Run")
plt.show()
```

```python
bf['winner'].value_counts()
```

```python
# make a barplot of top 3
plt.figure(figsize=(8,5))
colors =['green','black',"blue"]
plt.bar(list(bf['winner'].value_counts()[:3].keys()),list(bf['winner'].value_counts()[:3]),color=colors)
```

```python
# make a pie chart
plt.figure(figsize=(8,5))
plt.pie(list(bf['winner'].value_counts()),labels=list(bf['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```

```python
# lets make all steps for bating 2nd
bs = df[df['win_by_wickets']!=0]
```

```python
bs.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(bs['win_by_wickets'],bins=30)
plt.show()
```

```python
bs['winner'].value_counts()
```

```python
# make a barplot of top 3
plt.figure(figsize=(8,5))
colors =['green','black',"blue"]
plt.bar(list(bs['winner'].value_counts()[:3].keys()),list(bs['winner'].value_counts()[:3]),color=colors)
```

```python
# make a pie chart
plt.figure(figsize=(8,5))
plt.pie(list(bs['winner'].value_counts()),labels=list(bs['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```
