## *The Dataset colelcted from the source has follwoing description:*
`Description`
This dataset contains demographic information from the Pakistan Population Census conducted in 2017. It provides detailed population data at various administrative levels within Pakistan, including provinces, divisions, districts, and sub-divisions. The dataset also includes information on urban and rural populations, gender distribution, transgender individuals, sex ratios, population figures from the 1998 census, and annual growth rates.

`Features`
Province: The administrative provinces or regions of Pakistan where the census data was collected.

Division: The divisions within each province. Divisions are the second level of administrative divisions in Pakistan.

District: Districts within each division, representing larger administrative units.

Sub-Division: Sub-divisions or tehsils within each district, providing more localized data.

Area: The land area (in square kilometers) of each sub-division.

Urban Population 2017: The population of urban areas within each sub-division for the year 2017.

Rural Population 2017: The population of rural areas within each sub-division for the year 2017.

Male Population 2017: The male population within each sub-division for the year 2017.

Female Population 2017: The female population within each sub-division for the year 2017.

Transgender Population 2017: The population of transgender individuals within each sub-division for the year 2017.

Sex Ratio 2017: The sex ratio, calculated as the number of females per 1000 males, within each sub-division for the year 2017.

Population in 1998: The total population of each sub-division as recorded in the 1998 census.

Annual Growth Rate: The annual growth rate of the population in each sub-division, calculated as the percentage increase from 1998 to 2017.

```python
# How much Rural Population increased from 1998 - 2017?
pop_rural_1998 = df['POPULATION 1998 (RURAL)'].sum()
pop_rural_2017 = df['ALL SEXES (RURAL)'].sum()
pop_change_rural =  pop_rural_2017 - pop_rural_1998
pop_change_rural
print("The total Rural Population of Pakistan have been increased by 2017 as compared to 1998, and is as follows:",pop_change_rural)
```

```python
# How much Rural Population increased from 1998 - 2017?
pop_urban_1998 = df['POPULATION 1998 (URBAN)'].sum()
pop_urban_2017 = df['ALL SEXES (URBAN)'].sum()
pop_change_urban =  pop_urban_2017 - pop_urban_1998
pop_change_urban
print("The total Urbun Population of Pakistan have been increased by 2017 as compared to 1998, and is as follows:",pop_change_urban )
```

```python
# Population Change
percent_change = (pop_change_urban/pop_urban_1998)*100
print("The Urban Population of Pakistan increase by:", percent_change,"%, in 2017 from 1998.") #check if this is correct
```

```python

# Group the data by province and sum the population for each province
province_population = df.groupby('PROVINCE')[['ALL SEXES (RURAL)', 'ALL SEXES (URBAN)']].sum()

# Calculate total population (rural + urban) for each province
province_population['TOTAL POPULATION'] = province_population['ALL SEXES (RURAL)'] + province_population['ALL SEXES (URBAN)']

# Sort the provinces by total population in descending order
province_population_sorted = province_population.sort_values(by='TOTAL POPULATION', ascending=False)

# Plot the distribution of population across different provinces
plt.figure(figsize=(10, 6))
plt.bar(province_population_sorted.index, province_population_sorted['TOTAL POPULATION'], color='skyblue')
plt.title('Distribution of Population Across Provinces')
plt.xlabel('Province')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['ANNUAL GROWTH RATE (RURAL)'], label='Rural', marker='o')
plt.plot(df['ANNUAL GROWTH RATE (URBAN)'], label='Urban', marker='o')
plt.title('Population Growth Rate Variation Between Urban and Rural Areas')
plt.xlabel('Year')
plt.ylabel('Annual Growth Rate (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate population density for each division
df['POPULATION DENSITY'] = (df['ALL SEXES (RURAL)'] + df['ALL SEXES (URBAN)']) / df['AREA (sq.km)']

# Group the data by division and calculate the average population density for each division
division_population_density = df.groupby('DIVISION')['POPULATION DENSITY'].mean()

# Find the division with the highest population density
highest_density_division = division_population_density.idxmax()
highest_density_value = division_population_density.max()

print("The division with the highest population density is:", highest_density_division)
print("Population density:", highest_density_value)
```

```python

# Select columns relevant to population counts and district
population_columns = ['DISTRICT', 'POPULATION 1998 (RURAL)', 'POPULATION 1998 (URBAN)',
                      'ALL SEXES (RURAL)', 'ALL SEXES (URBAN)']

# Filter out rows where population data is missing
population_data_filtered = df.dropna(subset=population_columns)

# Calculate the total population in 1998 and the latest available year
population_data_filtered['TOTAL POPULATION 1998'] = population_data_filtered['POPULATION 1998 (RURAL)'] + population_data_filtered['POPULATION 1998 (URBAN)']
population_data_filtered['LATEST TOTAL POPULATION'] = population_data_filtered['ALL SEXES (RURAL)'] + population_data_filtered['ALL SEXES (URBAN)']

# Identify districts where the population has declined since 1998
population_declined_districts = population_data_filtered[population_data_filtered['LATEST TOTAL POPULATION'] < population_data_filtered['TOTAL POPULATION 1998']]['DISTRICT'].unique()

if len(population_declined_districts) > 0:
    print("Districts where the population has declined since 1998:")
    for district in population_declined_districts:
        print("-", district)
else:
    print("No districts where the population has declined since 1998.")
```
