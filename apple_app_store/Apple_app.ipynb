## Insights From Data

```python
# unique primary genres
unique_genres = df['Primary_Genre'].unique()
print("Unique primary genres of apps:", unique_genres)
```

```python
# average user rating
avg_user_rating = df['Average_User_Rating'].mean()
print("Average user rating of apps:", avg_user_rating)
```

```python
# most reviewed app and how many reviews does it have
most_reviewed_app = df.loc[df['Reviews'].idxmax()]
print("Most reviewed app:", most_reviewed_app['App_Name'])
print("Number of reviews:", most_reviewed_app['Reviews'])
```

```python
# average size of apps
avg_size_bytes = df['Size_Bytes'].mean()
print("Average size of apps (in bytes):", avg_size_bytes)
```

```python
# number of free and paid apps
num_free_apps = df[df['Free'] == True].shape[0]
num_paid_apps = df[df['Free'] == False].shape[0]
print("Number of free apps:", num_free_apps)
print("Number of paid apps:", num_paid_apps)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Free', data=df, palette='Set2')
plt.title('Count of Free vs. Paid Apps')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Paid', 'Free'])
plt.show()
```

```python
unique_content_ratings = df['Content_Rating'].unique()
print("Unique content ratings:", unique_content_ratings)
```

```python
# top 5 developers
top_developers = df['Developer'].value_counts().head(5)
print("Top 5 developers by number of apps:")
print(top_developers)
```

```python
# plot the top 5 developers
sns.barplot(x=top_developers.index, y=top_developers)
plt.title("Top 5 Developers by Number of Apps")
plt.xlabel("Developer")
plt.ylabel("Number of Apps")
plt.xticks(rotation=45)
plt.show()
```

```python
# most recent release date
most_recent_release = df['Released'].max()
print("Most recent release date:", most_recent_release)
```

```python
# average price of paid apps
avg_price_paid_apps = df[df['Free'] == False]['Price'].mean()
print("Average price of paid apps:", avg_price_paid_apps)
```

```python
# most common required iOS version
most_common_ios_version = df['Required_IOS_Version'].mode()[0]
print("Most common required iOS version:", most_common_ios_version)
```

```python
# distribution of app sizes across different content ratings
size_distribution_by_rating = df.groupby('Content_Rating')['Size_Bytes'].describe()
print("Distribution of app sizes across different content ratings:")
print(size_distribution_by_rating)
```

```python
# plot the distribution of app sizes across different content ratings
sns.boxplot(x='Content_Rating', y='Size_Bytes', data=df)
plt.title("Distribution of App Sizes Across Different Content Ratings")
plt.xlabel("Content Rating")
plt.ylabel("Size (in bytes)")
plt.xticks(rotation=45)
plt.show()
```

```python
# ratio of free to paid apps by primary genre
ratio_free_to_paid_by_genre = df.groupby('Primary_Genre')['Free'].value_counts(normalize=True).unstack().fillna(0)
print("Ratio of free to paid apps by primary genre:")
print(ratio_free_to_paid_by_genre)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Reviews', y='Average_User_Rating', data=df, alpha=0.5)
plt.title('Average User Rating vs. Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Average User Rating')
plt.show()
```

```python
# histogram of app sizes
plt.figure(figsize=(10, 6))
sns.histplot(df['Size_Bytes'], bins=20, kde=True)
plt.title('Histogram of App Sizes')
plt.xlabel('Size (Bytes)')
plt.ylabel('Frequency')
plt.show()
```
